```python
result = DataFrame(np.zeros((pivoted_csv.columns.size,pivoted_csv.columns.size)),columns=pivoted_csv.columns
                   ,index=pivoted_csv.columns)
def jaccard(x):
#     c = [a for i in x.columns.size if x.iloc[:,i] in x.iloc[:,i+1]
    for i in x.columns.size:
         for t in x.iloc[:,i]:
             if t in x.iloc[:,i+1]:
                result.iloc[i,i+1] = 'a'
                result.iloc[i+1,i] = 'a'               
    return result
```
